## 🛠️ Paso 1: Scatter Plot Básico (Contínuo vs Contínuo)

**Objetivo:** Analizar si existe una relación visual directa entre la edad de las personas y los cargos médicos que se les cobran.

**Instrucciones:**
1. Crea un gráfico de dispersión (Scatter Plot) donde el eje X sea la edad (`age`) y el eje Y sean los cargos (`charges`).
2. Utiliza la librería Seaborn con su función por defecto para este tipo de gráficos.
3. Observa si ves alguna tendencia ascendente, descendente o agrupamientos.

```python
# Tu código aquí
# creamos el estilo de los gráficos de seaborn
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
# gráfico de dispersión con seaborn de los cargos médicos vs la edad
df.plot.scatter(
    x='age',
    y='charges'
)
plt.title('Cargos Médicos vs Edad')
```

## 🛠️ Paso 2: Análisis multivariable usando color (Hue)

**Objetivo:** Descubrir si el hecho de fumar afecta significativamente a los cargos médicos, manteniendo la relación con la edad.

**Instrucciones:**
1. Utiliza el mismo tipo de gráfico que en el paso anterior.
2. Añade un parámetro para colorear los puntos dependiendo de la clase categórica si la persona fuma o no (`smoker`).
3. Fíjate en cómo se separan visualmente los grupos.

```python
# Tu código aquí
# como dice el enunciado, el gráfico de dispersión con seaborn de los cargos médicos vs la edad por fumador
plot = sns.catplot(
    data=df,
    x="age",
    y="charges",
    hue="smoker",
    kind="strip",
    height=6,
    aspect=1.5,
    alpha=0.6,
    dodge=True,
)

plot.set_xticklabels(rotation=45)
plt.title('Cargos médicos vs edad por fumador')
plt.tight_layout()
```

## 🛠️ Paso 3: Scatter Categórica (Categórico vs Numérico)

**Objetivo:** Agrupar y comparar visualmente una variable numérica en base a diferentes categorías, estudiando la densidad y dispersión.

**Instrucciones:**
1. Vamos a analizar la variable índice de masa corporal (`bmi`) y compararla según la región de residencia (`region`).
2. Utiliza la función orientada a variables categóricas de dispersión de Seaborn (`stripplot` o `swarmplot`).
3. Pon la región (`region`) en el eje X y el IMC (`bmi`) en el eje Y.

```python
# Tu código aquí
# vamos ahora a graficar la distribución del IMC por región con un gráfico de dispersión con seaborn
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x='region', y='bmi', alpha=0.7)
plt.title('Distribución del IMC por Región')
plt.xlabel('Región')
plt.ylabel('Índice de Masa Corporal (BMI)')
plt.show()
```

## 🛠️ Paso 4: Ajustes Finales y Customización

**Objetivo:** Hacer que un gráfico sea presentable para un reporte, controlando las opciones estéticas.

**Instrucciones:**
1. Vamos a crear un nuevo gráfico comparando IMC (`bmi`) en el eje X frente a `charges` en el eje Y, y coloréalo según `smoker`.
2. Asigna un tamaño a la figura usando `plt.figure(figsize=(10, 6))`.
3. Dale transparencia a los puntos (parámetro `alpha`).
4. Ponle un título informativo al gráfico usando `plt.title()`.
5. Pon etiquetas entendibles en los ejes X e Y usando `plt.xlabel()` y `plt.ylabel()`.

```python
# Tu código aquí
# por último, vamos a graficar los cargos médicos vs el IMC, diferenciando por 
# fumadores y no fumadores con un gráfico de dispersión con seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bmi', y='charges', hue='smoker', alpha=0.7)
plt.title('Cargos Médicos según IMC y Hábito de Fumar')
plt.xlabel('Índice de Masa Corporal (BMI)')
plt.ylabel('Cargos Médicos')
plt.show()
```
